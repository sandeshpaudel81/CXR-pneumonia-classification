```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from glob import glob
```

```python
train_path = '/content/chest_xray/train'
val_path = '/content/chest_xray/val'
test_path = '/content/chest_xray/test'
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 10,
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory(test_path,
                                            target_size = (224, 224),
                                            batch_size = 10,
                                            class_mode = 'categorical')
```

```python
IMAGE_SIZE = [224, 224]
```

```python
base_model = tf.keras.applications.vgg16.VGG16(
    include_top=False,
    input_shape=IMAGE_SIZE+[3],
    weights = 'imagenet'
)
for layer in base_model.layers:
    layer.trainable = False

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit_generator(
    training_set,
    validation_data=test_set,
    epochs=30,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(range(30), acc, label='Training Accuracy')
plt.plot(range(30), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(range(30), loss, label='Training Loss')
plt.plot(range(30), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
test_normal_images = glob('/content/chest_xray/val/NORMAL/*jpeg')
```

```python
test_pneumonia_images = glob('/content/chest_xray/val/PNEUMONIA/*jpeg')
```

```python
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
```

```python
def test_accuracy(normal, pneumonia):
    pred = []
    for i in range(len(normal)):
        img = image.load_img(normal[i], target_size=(224,224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        img_data = preprocess_input(x)
        classes=model.predict(img_data)
        print(classes)
        if (classes[0][1]>0.3):
            pred.append(1)
        else:
            pred.append(0)
    for i in range(len(pneumonia)):
        img = image.load_img(pneumonia[i], target_size=(224,224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        img_data = preprocess_input(x)
        classes=model.predict(img_data)
        print(classes)
        if (classes[0][1]>0.3):
            pred.append(1)
        else:
            pred.append(0)
    return pred
```

```python
def get_test_labels(normal, pneumonia):
    labels = []
    for i in range(len(normal)):
        labels.append(0)
    for i in range(len(pneumonia)):
        labels.append(1)
    return labels

# import cv2
# def process_image(img_path):
#     img = cv2.imread(img_path)
#     img = cv2.resize(img, (224, 224))
#     img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#     img = img/255.0
#     img = np.reshape(img, (224,224,3))
#     return img
```

```python
test_labels = get_test_labels(test_normal_images, test_pneumonia_images)
test_predictions = test_accuracy(test_normal_images, test_pneumonia_images)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_labels, test_predictions)
accuracy
```
